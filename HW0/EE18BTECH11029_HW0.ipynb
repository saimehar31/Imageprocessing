```python

# 2_9) Solution

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from numpy import asarray
img = mpimg.imread('5.1.10.tiff') 
data = asarray(img)


t = np.zeros([256,256]);


#------------------------------------------------# This is for existing all bits
for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]

plt.figure()
plt.imshow(t,cmap='gray') 

#------------------------------------------------# This is for quantizing 1 bit
for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>1

plt.figure()
plt.imshow(t,cmap='gray') 


#------------------------------------------------# This is for quantizing 2 bit
for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>2

plt.figure()
plt.imshow(t,cmap='gray') 

#------------------------------------------------# This is for quantizing 3 bit

for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>3

plt.figure()
plt.imshow(t,cmap='gray') 

#------------------------------------------------# This is for quantizing 4 bit

for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>4

plt.figure()
plt.imshow(t,cmap='gray') 

#------------------------------------------------# This is for quantizing 1 bit

for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>5

plt.figure()
plt.imshow(t,cmap='gray') 

#------------------------------------------------# This is for quantizing 5 bit

for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>6

plt.figure()
plt.imshow(t,cmap='gray') 
 
#------------------------------------------------# This is for quantizing 6 bit
for i in range(256):
  for j in range(256):
    t[i][j] = img[i][j]>>7

plt.figure()
plt.imshow(t,cmap='gray') 
#------------------------------------------------# This is for quantizing 7 bit

```
