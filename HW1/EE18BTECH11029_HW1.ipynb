```python
import matplotlib.image as mpimg 
import math
import matplotlib.pyplot as plt 
from numpy import asarray                                           
import numpy as np
I = mpimg.imread('input.jpg')

J = np.full( ( len(I), len(I[0]) ) , 0)


for i in range (len(I)):
    for j in range(len(I[i])):
            t[i][j]=int(((int(I[i][j][0])+int(I[i][j][1])+int(I[i][j][2]))/int(3)));

h=np.full(256,0)
"""
Calculating the frequency of elements and then normalizing it and then again finding the CDF of it and
then applying FSCS operation
"""

for i in range (len(I)):
    for j in range (len(I[i])):
        h[int(t[i][j])]+=1;
        
p=np.zeros(256)
        
for k in range (len(h)):
    p[k]=h[k]/(len(t)*len(t[0]))
    
cdfval=np.zeros(256)

ans=0

for i in range(len(p)):
    ans=ans+p[i];
    cdfval[i]=ans;
    

J = np.full( ( len(t), len(t[0])) , 0)

for i in range(len(t)):
    for j in range(len(t[0])):
        J[i][j]=cdfval[math.ceil(t[i][j])]
        

"""
Applying FSCS operation
"""
B=np.max(J)
A=np.min(J)

K=256;

P=(K-1)/(B-A)
L= -A *((K-1)/(B-A))

F=np.zeros([len(J),len(J[0])])

for x in range(len(I)):
    for y in range(len(I[x])):
        F[x][y]=(P * J[x][y]) + L ;

plt.hist(J,bins=8)   
```
